```python
Jugadores = {}
stats = {}
datastat = data.columns
datastat = datastat.drop("Player")

for i in range(len(data["Player"])):
    for stat in datastat:
        stats.update({stat:data[stat][i]})
    Jugadores.update({data["Player"][i]:stats.copy()})
```

```python
p=[]
p.append(data[(data['Control']>=70) & (data['Control']<80)]['Control'].count())
p.append(data[(data['Control']>=80) & (data['Control']<90)]['Control'].count())    
p.append(data[(data['Control']>=90) & (data['Control']<100)]['Control'].count())
p.append(data[(data['Control']>=100) & (data['Control']<110)]['Control'].count())
p.append(data[(data['Control']>=110) & (data['Control']<120)]['Control'].count())
p.append(data[(data['Control']>=120) & (data['Control']<130)]['Control'].count())  
p.append(data[data['Control']>=140]['Control'].count())
p
```
